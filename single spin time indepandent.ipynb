```python
%matplotlib inline
from qutip import *
from scipy import *
import numpy as np
import matplotlib.pyplot as plt

vec_eigen = np.zeros((3,200,2))
vec_h = np.linspace(-5,5,200)

def findeigen(a,vec_eigen):
    
    for j,k in enumerate(vec_h):
        H = a * sigmax() + k * sigmaz()
        vec_eigen[j] = H.eigenenergies()
        
list_a = [0, 1, 2]
for j, a in enumerate(list_a):
    findeigen(a, vec_eigen[j])

plt.figure(1)
plt.plot(vec_h,vec_eigen[0],vec_h,vec_eigen[1],vec_h,vec_eigen[2])
```
